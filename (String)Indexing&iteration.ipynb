# Indexing and slicing

```python

# indexing - position ---- access character from given string
s="welcome to bhopal"
print(s[1])
#slicing - extract substring from given string
# strobj[start:stop:step]
# start - 0
# STOP - end of the string - given condition -- n-1
# step - +1
# slicing
print(s[::-1])  #reverse string
print(s[1::3])
print(s[-1::-2])
print(s[1:6:2])
print(s[0:7:])
print(s[-1:-14:-2])
```

```python
# iteration
# direct and indirect
s="decode"
for i in s:
    print(i)
```

```python
# indirect
s="decode"
for i in range(len(s)):
    print(i,s[i])
```

```python
# wapp  to print reverse of any given string
s=input("enter any string : ")
rev=""
for i in s:
    rev=i+rev
print(rev)       #with direct iteration
print(s[::-1])   #with slicing
```

```python
# wapp to check whether a given is palindrome or not
s=input("enter any string : ")
rev=""
for i in s:
    rev=i+rev
    print
if rev==s:
    print(rev,"given string is palindrome")
else:
    print(rev,"given string is not palindrome")
```

```python
s=input("enter any string : ")
if s==s[::-1]:
    print(s,"given string is palindrome")
else:
    print(s,"given string is not palindrome")
```

```python
# count() - return occurence of particular character/substring in a given string
# strobj.count(string/character,[start],[stop])
s="welcome to bhopal"
print(s.count("o"))
print(s.count("o",5))
print(s.count("o",5,10))
print(s.count("o",5,9))
print(s.count("decode"))
print(s.count("bhopal"))
```

```python
# strip()
# strobj.strip(string)
s="...?#rrrdecodett,,$,,/"
print(s.strip(".?#rt,$/"))
```

```python
# index() - return position
# strobj.index(char/substring,[start],[stop])
s="welcome to bhopal"
print(s.index("o"))
print(s.index("o",5))
print(s.index("bhopal"))
print(s.index("o",5,9))
```

```python
# find() - return position
# strobj.index(char/substring,[start],[stop])
s="welcome to bhopal"
print(s.find("o"))
print(s.find("o",5))
print(s.find("bhopal"))
print(s.find("o",5,9))
```

```python
# replace
# strobj.replace(oldstring,newstring,[count])
s="welcome to bhopal"
print(s.replace("bhopal","decode"))
print(s.replace("o","a"))
```

```python
# split()
# strobj.split([seperator],[maxsplit])
# by default - whitespace , maxsplit= max no of seperator
# split always return list of strings.
s="welcome to city of lakes"
print(s.split())
print(s.split(" ",1))
print(s.split(" "))
```
