```python
#if else
num=int(input("enter any number: "))
if num%2==0:
    print(num,"is an even number")
else:
    print(num,"is an odd number")
```

```python
science=int(input("enter marks in science :"))
lang=int(input("enter markse in language :"))
maths=int(input("enter marks in maths :"))
obtained_marks=science+lang+maths
per=(obtained_marks/300)*100
print("obtained marks :",obtained_marks)
print("percentage :",per)
if per>=65:
    print("1st division ")
elif per>=50:
    print("2nd division ")
elif per>=33:
    print("3rd division ")
else:
    print("fail")
    
```

```python
a1=int(input("enter length of 1st side"))
a2=int(input("enter length of 2nd side"))
a3=int(input("enter length of 3rd side"))
if a1!=0 and a2!=0 and a3!=0:
    if  a1+a2>a3 and  a1+a3>a2 and a2+a3>a1:
        if a1==a2 and a1==a3:
            print("equilateral triangle ")
        elif a1==a2 or a2==a3:
            print("isosceles triangle ")
        else:
            print("scalene triangle ")
    else:
        print("values incorrect")
else:
    print("invalid input")
```

```python
year=int(input("enter any year"))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year,"century leap year")
        else:
            print(year,"century not a leap year")
    else:
        print(year,"non century leap year")
else:
    print(year,"non century not a leap year")
```
