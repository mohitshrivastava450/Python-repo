```python
#LIST
l=[]
print(l)
print(type(l))
```

```python
l=[10,20,30,"hello",10.5]
print(l)
print(type(l))
print(l[0])   #indexed
l[0]=1000     #mutable
print(l)
```

```python
#indexing and slicing
l=[10,20,30,"hello",10.5,1000,200,300,100]
print(l[3])
print(l[3][1])
print(l[-3])
print(l[::-1])
print(l[::])
print(l[:6:2])
print(l[-1:-7:-2])
print(l[-1:7:-2])
```

```python
#direct iteration
l=[10,20,30,"hello",1000,10.5,11] 
for i in l:
    print(i)
```

```python
#indirect iteration
l=[10,20,30,"hello",1000,10.5,11] 
for i in range(len(l)):
    print(l[i])
```

```python
#function
# len()
# max()
# min()
l=[10,20,30,"hello",1000,10.5,11] 
print(len(l))
count=0
for i in l:
    count= count+1
print(count)    
```

```python
#max()
# min()
l=[10,20,30,1000,10.5,11]
print(max(l))
print(min(l))
```

```python
l=[10,20,30,1000,10.5,11]
max_value=l[0]
for i in l:
    if i>max_value:
        max_value=i
print(max_value)        
```

```python
l=[10,20,30,1000,10.5,11]
min_value=l[0]
for i in l:
    if i<min_value:
        min_value=i
print(min_value)   
```

```python
l=[10,20,30,1000,10.5,11]
print(sum(l))
sum=0
for i in l:
    sum=sum+i
print(sum)    
```
