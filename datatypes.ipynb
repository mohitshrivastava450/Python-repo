# string -ordered sequence of characters

## singleline string

```python
#single line denoted by '' or "" and
#multiline string denoted by ''' '''
#indexed and immutable
s="" #empty string
print(s)
print(type(s))
```

## string immutable error

```python
s="decode"
print(s)
print(type(s))
print(s[0])
s[0]="z"
print(s)
```

## multiline string

```python
s='''hello
welcome 
to bhopal'''
print(s)
```

# List

```python
# list -collection of element
# it is denoted by [] with csv
# indexed,ordered,mutable and heterogenours in nature
l=[] #empty list
print(l)
print(type(l))
```

```python
l=[10,20,30,"decode",10.5]
print(l)
print(l[3]) #indexed
l[0]="bhopal"  #mutable
print(l)
```

# Tuple

## empty tuple

```python
# tuple - collection of element'
# denoted by () with csv
# ordered,indexed,immutable and heterogenous
t=() #empty tuple
print(t)
print(type(t))
```

## single value tuple

```python
# single value tuple
t=('10.5',)
print(t)
print(type(t))
```

## multi value tuple

```python
#multi value tuple
t=(10,20,300,"decode",10.5)
print(t)
print(t[2])
t[2]=400
print(t)
```

# Dictionary

```python
# dictionary
# collection of datavalue
# it is denoted by {k:v} with key:value pair
# ordered , unindexed and mutable
# keys --unique
# values -- duplicate
d={}  # empty dictionary dict
print(d)
print(type(d))
```

```python
d={"name":"abc","subject":"DA","fees":30000}
print(d)
print(d["name"])
d["name"]="def"
print(d)
```

```python
d={}
d["name"]="def"
print(d)
```

```python
d={"name":["def","abc","pqr"],"subject":["DA","DA","DS"]}
print(d)
print(d["name"])
```

# set

```python
#set - collection of element
# denoted by {} with csv
# unordered,unindexed ,mutable , heterogenous in nature
# stored unique value
s=set() #set construction  s=() does not create an empty set
print(s)
print(type(s))
```

```python
s={10,10,10,20,10,100,200,"decode","h"} #display duplicate values only one time 
print(s)  
print(type(s))
s.add(1000)
print(s)
```
