```python
# get
#dictobj.get(key):
d={"name":["abc","def","pqr"],"courses":["DA","DS","DE"],"fees":[25000,35000,30000]}
print(d.get("name"))
for i in d.get("name"):
    print(i)
```

```python
# keys()
# dictobj.keys()
d={"name":["abc","def","pqr"],"courses":["DA","DS","DE"],"fees":[25000,35000,30000]}
print(d.keys())
for i in (d.keys()):
    print(i)
```

```python
# values()
# dictobj.values()
d={"name":["abc","def","pqr"],"courses":["DA","DS","DE"],"fees":[25000,35000,30000]}
print(d.values())
for i in (d.values()):
    print(i)
```

```python
# items()
# dictobj.items()
d={"name":["abc","def","pqr"],"courses":["DA","DS","DE"],"fees":[25000,35000,30000]}
print(d.items())
for i in (d.items()):
    print(i)
```

```python
d={"name":["abc","def","pqr"],"courses":["DA","DS","DE"],"fees":[25000,35000,30000]}
print(d.items())
for i,j in (d.items()):
    print(i,j)
```

```python
l=[1,1,3,3,3,3,3,4,6,6,6]
d={}
for i in l:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
print(d)
max_value=max(d.values())
print(max_value)
mode=[i for i,j in d.items() if j==max_value]
print(mode)
```

```python
#wapp to print lcm and hcf of given numbers take 2 user input
a=int(input("Enter 1st number: "))
b=int(input("Enter 2nd number: "))
l1=[]
l2=[]
hcf=0
lcm=0
for i in range(1,a+1):
    if a%i==0:
        l1.append(i)
for i in range(1,b+1):
    if b%i==0:
        l2.append(i)
for i in l1:
    for j in l2:
        if i==j:
            hcf=i
            #print(hcf)
for i in l1:
    for j in l2:
        if i%j==0:
            lcm=i
        elif j%i==0:
            lcm=j
        else:
            lcm=i*j
print(l1)
print(l2)
print("HCF:",hcf)
print("LCM:",lcm)
```

```python
# pop()
# dictobj.pop(key)
d={"name":["abc","def","pqr"],"courses":["DA","DS","DE"],"fees":[25000,35000,30000]}
print(d.pop("name"))
print(d)
```

```python
d={"name":["abc","def","pqr"],"courses":["DA","DS","DE"],"fees":[25000,35000,30000]}
print(d["name"].pop())
print(d)
```

```python
# popitem()
# dictobj.popitem
# return key and value pair in a tuple
d={"name":["abc","def","pqr"],"courses":["DA","DS","DE"],"fees":[25000,35000,30000]}
print(d.popitem())
print(d)
```

```python
# del
d={"name":["abc","def","pqr"],"courses":["DA","DS","DE"],"fees":[25000,35000,30000]}
del d["name"]
print(d)
```

```python
# clear()
# dictobj.clear()
d={"name":["abc","def","pqr"],"courses":["DA","DS","DE"],"fees":[25000,35000,30000]}
d.clear()
print(d)
```

```python
# update
# dictobj.update(dictionary)
d={"name":["abc","def","pqr"],"courses":["DA","DS","DE"],"fees":[25000,35000,30000]}
d.update({"city":["jbp","bpl","gwl"]})
print(d)
```

```python
d={"name":["abc","def","pqr"],"courses":["DA","DS","DE"],"fees":[25000,35000,30000]}
d.update(city=["jbp","bpl","gwl"])
print(d)
```

```python
# dict()
d=dict(name=["abc","def","pqr"],city=["jbp","bpl","gwl"])
print(d)
```

```python
# dictionary comprehension - to create a new dictionary in a concise way.
# (k:v looping statement conditional statement)
```

```python
d={}
for i in range(2,11,2):
    d[i]=i**2
print(d)
print({i:i**2 for i in range(2,11,2)})
```

```python
d={}
l=[1,2,3,4,5]
for i in l:
    d[i]=i**3
print(d)
print({i:i**3 for i in l})
```

```python
# problem - map numbers to "even" or "odd"
# but only for number greater than 3 and else then 10?
# {4:"even",5:"odd",6:"even",7:"odd",8:"even",9:"odd"}
number=range(10)
d={}
for i in number:
    if i>3:
        if i%2==0:
            d[i]="even"
        else:
            d[i]="odd"
print(d)
print({i:("even" if i%2==0 else "odd") for i in number if i>3})
```
