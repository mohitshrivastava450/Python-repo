```python
# wapp to print first n even and first odd numbers
n=int(input("enter a number : "))
for i in range(1,(2*n)+1):
    if i%2==0:
        print(i,"even number ")
    else:
        print(i,"odd number ")
```

```python
# wap to print sum of first n natural numbers
n=int(input("enter any number : "))
sum_natural=0
for i in range(1,n+1):
    sum_natural=sum_natural+i
print(sum_natural)
```

```python
# wap to print sum of first n even and n odd numbers
n=int(input("enter a number"))
sum_even=0
sum_odd=0
n1=0
n2=0
for i in range(1,(2*n)+1):
    if i%2==0:
        n1=i
        sum_even=sum_even+n1
    else:
        n2=i
        sum_odd=sum_odd+n2
print("sum of even numbers is ",sum_even)
print("sum of odd numbers is ",sum_odd)
```

```python
# wap to print count of even and odd numbers between 1000 and 2000 
n1=int(input("enter range start value: "))
n2=int(input("enter range end value : "))
a1=0
a2=0
for i in range(n1,n2+1,1):
    if i%2==0:
        a1=a1+1
    else:
        a2=a2+1
print("Even number count : ",a1)
print("Odd number count : ",a2)
        
```
