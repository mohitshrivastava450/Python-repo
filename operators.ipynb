# operators

## arithmetic operators

```python
# operators
# arithmetic operators
a=10
b=2
print("addition ",a+b)
print("substraction ",a-b)
print("multiplication ",a*b)
print("simple division ",a/b)
print("floor division ",a//b)
print("modulus ",a%b)
print("exponential power ",a**b)
```

## assignment operators  =,+=,-=,*=,/*,//=,**==

```python
# assignment operators  =,+=,-=,*=,/*,//=,**==
a=10
b=8
c=a+b
a+=b  #a=a+b
a-=10  #a=a-10
a*=5  # a=a*5
a//=5  #a=a//5
b**=10  #b=b*10
print(a)
print(b)
print(c)
```

## comparision operator ==,!=,>=,<=,>,<

```python
# comparision operator ==,!=,>=,<=,>,<
a=10
b=8
print(a==b)
print(a!=b)
print(a>=b)
print(a<=b)
print(a>b)
print(a<b)
```

## identity operator is ,is not

```python
# identity operator is ,is not
#id() - return memory address of an object
a=10
b=10
print(id(a))
print(id(b))
print(a is b)
print(a is not b)
```

## membership operator in ,not in

```python
#membership operator in ,not in
a="hello"
print("he" in a)
print("h" not in a)
l=(10,20,"hello")
print("hel" in l)
```

## bitwise

```python
# bitwise
# &- bitwise and
# |- bitwise or
# bin()
a=10
b=8
print(a,bin(a))
print(b,"",bin(b))
print(bin(a&b))
print(a&b)
print()
print(bin(a|b))
print(a|b)
print()
print(bin(a^b))
print(a^b)
```
