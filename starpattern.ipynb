# Program to print star matrix by taking user input

```python
n=int(input("enter any number"))
for i in range(n):
    for j in range(n):
       print("*",end=" ")
    print()
```

```python
n=int(input("enter any number"))
for i in range(n):
    for j in range(i+1):
        print("*",end=" ")
    print()
```

```python
n=int(input("enter any number"))
for i in range (n):
    for j in range (n-i):
        print("*",end=" ")
    print()
```

```python
n=int(input("enter any number"))
for i in range(n):
    for j in range(n-i-1):
        print("$",end=" ")
    for j in range(i+1):
        print("*",end=" ")
    print()
```
