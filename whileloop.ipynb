# while loop

```python
n=int(input("enter a number"))
i=1
while i<=n:
    print(i,"decode data")
    i=i+1
```

```python
#wap to print sum of first n natural number
n=int(input("enter any number: "))
sum_natural=0
i=1
while i<=n:
    sum_natural=sum_natural+i
    i=i+1
print(sum_natural)
```

```python
#wap to print sum of first n even number & odd number
n=int(input("enter any number: "))
i=1
sum_even=0
sum_odd=0
while i<=2*n:
    if i%2==0:
        sum_even=sum_even+i
    else:
        sum_odd=sum_odd+i
    i=i+1
print(sum_even)
print(sum_odd)
```

```python
#wap to print factorial of n number
n=int(input("enter any number: "))
i=1     #start
fact=1
while i<n+1:    # stop
    fact=fact*i  #logic
    i=i+1       #step
print(fact)
```
